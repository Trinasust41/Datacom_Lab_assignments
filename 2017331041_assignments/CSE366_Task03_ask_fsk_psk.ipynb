```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
time=np.arange(0,1,.001)
time
```

```python
digital_data=[0,1,0,1,1]
digital_signal=np.repeat(digital_data,200)
```

```python
def analog(A,f,time,phase):
  y=A*np.sin(2.0*np.pi*f*time+phase)
  return y
```

ASK

```python


plt.subplot(3, 1, 1)
plt.plot(time,digital_signal,'r')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Message Signal')


cf=50.0
cs=analog(1,cf,time,0)
plt.subplot(3, 1, 2)
plt.plot(time,cs,'g')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Carrier Signal')


analog_signal=[]
for i in range(len(time)):
  analog_signal.append(analog(1,cf,time[i],0)*digital_signal[i])
plt.subplot(3, 1, 3)
plt.plot(time,analog_signal)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('ASK Signal')
fig = plt.gcf()
fig.set_size_inches(18,16)
plt.show()
```

#PSK

```python
plt.subplot(3, 1, 1)
plt.plot(time,digital_signal,'r')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Message Signal')



cf=60.0
cs=analog(3,cf,time,0)
plt.subplot(3, 1, 2)
plt.plot(time,cs,'g')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Carrier Signal')



analog_signal_2=[]


for i in range(len(time)):
    if(digital_signal[i]==1):
         analog_signal_2.append(analog(3,cf,time[i],0))
    else:
        analog_signal_2.append(analog(3,cf,time[i],180))


plt.subplot(3, 1, 3)
plt.plot(time,analog_signal_2)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('PSK Signal')
fig = plt.gcf()
fig.set_size_inches(18,16)
plt.show()
```

#FSK

```python
plt.subplot(4, 1, 1)
plt.plot(time,digital_signal,'r')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Message Signal')



cf_1=30.0
cs=analog(3,cf_1,time,0)
plt.subplot(4, 1, 2)
plt.plot(time,cs,'g')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Carrier Signal no 1')

cf_2=60.0
cs=analog(3,cf_2,time,0)
plt.subplot(4, 1, 3)
plt.plot(time,cs,'y')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Carrier Signal no 2')



analog_signal_3=[]


for i in range(len(time)):
    if(digital_signal[i]==0):
         analog_signal_3.append(analog(3,cf_1,time[i],0))
    else:
        analog_signal_3.append(analog(3,cf_2,time[i],0))


plt.subplot(4, 1, 4)
plt.plot(time,analog_signal_3)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('PSK Signal')
fig = plt.gcf()
fig.set_size_inches(18,22)
plt.show()
```
