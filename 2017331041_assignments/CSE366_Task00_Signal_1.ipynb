```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# in np.arage the point will be repeated after a certain value 
t0 = np.arange(0.0, 2.0, 0.05)
t0
```

```python
# in np.linespace there will be certain numbers of points
t = np.linspace(0, 2.0, 100)
t
```

#Equation

$$x(t) = A\sin(2\pi f t + phase)$$

```python
def create_sinusoidal(A, f, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x
```

**A Sinusoidal Signal $x(t)$ and 3 different plotting in subplots**

```python
x = create_sinusoidal(2,1)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('Different subplots of a sinusoidal signal', fontsize=15)



ax[0, 0].plot(t,x)
ax[0, 0].set_title('continuous plot', fontsize = 15)

ax[0, 1].stem(t,x)
ax[0, 1].set_title('stem plot', fontsize = 15)

ax[1,0].scatter(t,x)
ax[1,0].set_title('scatter plot', fontsize = 15)

#ax[1,1].plot(t,x)
#ax[1,1].set_title('continuous plot', fontsize = 15)

ax[-1,-1].axis('off')

#for a in ax.flat:
 #    a.set(xlabel='t', ylabel='x(t)')

plt.show()
```

**Three Sinusoidal Signal $x1(t),x2(t),x3(t)$ then combine and plot the signal**

```python
x1 = create_sinusoidal(1,1)
x2 = create_sinusoidal(2,5)
x3 = create_sinusoidal(3,5)
#x4 = create_sinusoidal(3,3)
#x5 = create_sinusoidal(5,10)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('3 Sinusoidal signal and a combined signal', fontsize=15)
#
ax[0, 0].plot(t,x1, 'r')
ax[0, 0].set_title('x1(t)', fontsize = 15)

ax[0, 1].plot(t,x2, 'g')
ax[0, 1].set_title('x2(t)', fontsize = 15)

ax[1,0].plot(t,x3, 'b')
ax[1,0].set_title('x3(t)', fontsize = 15)
 
#ax[0,2].plot(t,x4, 'b')
#ax[0,2].set_title('x4(t)', fontsize = 15)

ax[1,1].plot(t, x1 + x2 + x3,'y')
ax[1,1].set_title('y(t) = x1(t) + x2(t) + x3(t)', fontsize = 15)
plt.show()
```
