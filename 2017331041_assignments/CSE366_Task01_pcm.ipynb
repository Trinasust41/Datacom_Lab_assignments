```

import matplotlib.pyplot as plt
import numpy as np
import math
```

```



plt.figure(figsize=(15,15)) 

lim = 1.0
#analog signal

a=1.0
f=5.0

d=-np.pi/2.0


t=np.linspace(-lim,lim,1500)


x_t=a*np.cos(2.0*np.pi*f*t+d)


plt.subplot(4,1,1)
plt.plot(t,x_t)

#sampling

Fs=110.0
T=1.0/Fs

nT= np.arange(-lim,lim,T)

x_nT=a*np.cos(2.0*np.pi*f*nT+d)

plt.subplot(4,1,2)
plt.stem(nT,x_nT)


#quantizing

bits=3;
levels=2**(bits-1)
step=1.0/levels

xq_n=np.round(x_nT/step)*step

plt.subplot(4,1,3)
plt.stem(nT,xq_n)


#Digital Signal
plt.subplot(4,1,4).set_title("Digital Signal")
xq_n = np.repeat(xq_n,8*levels)
plt.axhline(y=0,color='r',linestyle='-')
plt.plot(xq_n,'b')
```
