# Basic Signal

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0.0,2.0,0.00001)
t
```

#Equation

$$x(t) = A\sin(2\pi f t + phase)$$

```python
def create_sinusoidal(t, A, f, phase = 0):
  x = A * np.sin(2.0 * np.pi* f * t + phase)
  return x
```

```python
x1 = create_sinusoidal(t, 1, 5)
x2 = create_sinusoidal(t, 2, 10)
x3 = create_sinusoidal(t, 3, 15)
y = x1 + x2 + x3
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('3 Sinusoidal Signals and 1 Composite Signal', fontsize=15)
ax[0,0].plot(t, x1,'r')
ax[0,0].set_title('Signal-1', fontsize = 15)

ax[0,1].plot(t, x2,'g')
ax[0,1].set_title('Signal-2', fontsize = 15)

ax[1,0].plot(t, x3,'b')
ax[1,0].set_title('Signal-3', fontsize = 15)

ax[1,1].plot(t, y,'y')
ax[1,1].set_title('Signal-1 + Signal-2 + Signal-3', fontsize = 15)

plt.show()
```

```python
fs = 10
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 1, 5) + create_sinusoidal(n, 1, 10) + create_sinusoidal(n, 1, 15)
plt.title('fs = 10', fontsize = 15)
plt.stem(n,y,'r')
plt.show()

fs = 40
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 1, 5) + create_sinusoidal(n, 1, 10) + create_sinusoidal(n, 1, 15)
plt.title('fs = 20', fontsize = 15)
plt.stem(n,y,'g')
plt.show()

fs = 60
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 1, 5) + create_sinusoidal(n, 1, 10) + create_sinusoidal(n, 1, 15)
plt.title('fs = 50', fontsize = 15)
plt.stem(n,y,'b')

plt.show()
```
