#Amplitude modulation

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Carrier wave c(t)=A_c*cos(2*pi*f_c*t)
#Modulating wave m(t)=A_m*cos(2*pi*f_m*t)
#Modulated wave s(t)=A_c[1+mu*cos(2*pi*f_m*t)]cos(2*pi*f_c*t)

A_c = float(input('Enter carrier amplitude: '))
f_c = float(input('Enter carrier frquency: '))
A_m = float(input('Enter message amplitude: '))
f_m = float(input('Enter message frquency: '))
modulation_index = float(input('Enter modulation index: '))

t = np.arange(0, 50, 0.1)

carrier = A_c*np.sin(2*np.pi*f_c*t)
modulator =A_m*np.cos((2*np.pi*f_m*t)+0)
product = A_c*(1+modulation_index*np.cos(2*np.pi*f_m*t))*np.sin(2*np.pi*f_c*t)

plt.subplot(3,1,1)
plt.title('Amplitude Modulation')
plt.plot(modulator,'g')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(carrier, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(product, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('AM signal')

plt.subplots_adjust(hspace = 1)
plt.rc('font', size=20)
fig = plt.gcf()
fig.set_size_inches(20,9)
plt.show()

```

#Frequency modulation


```python
import numpy as np
import matplotlib.pyplot as plt

modulator_frequency = 4.0
carrier_frequency = 40.0
modulation_index = 1.0

time = t=np.arange(0, 1.0, 0.0001)
modulator=np.sin((2*np.pi*modulator_frequency*time))
carrier = 2*np.cos(2*np.pi*carrier_frequency*time); 
product =2*np.cos(2. * np.pi * (carrier_frequency * time + modulation_index * np.sin(2*np.pi*modulator_frequency*time)))



plt.subplot(3, 1, 1)
plt.title('Frequency Modulation')
plt.plot(modulator,'g')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')
plt.subplot(3, 1, 2)
plt.plot(carrier,'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')
plt.subplot(3, 1, 3)
plt.plot(product,'purple')
plt.ylabel('Amplitude')
plt.xlabel('Output signal')

plt.subplots_adjust(hspace = 1)
plt.rc('font', size = 30)
fig = plt.gcf()
fig.set_size_inches(16,9)
plt.show()


```

#Phase Modulation

```python
import numpy as np
import matplotlib.pyplot as plt

A_c = float(input('Enter carrier amplitude: '))
f_c = float(input('Enter carrier frequency: '))
A_m = float(input('Enter message amplitude: '))
f_m = float(input('Enter message frequency: '))
modulation_index = float(input('Enter modulation index: '))

t = np.arange(0, 3.0, 0.0001)

carrier = A_c*np.sin(2*np.pi*f_c*t + 0)
modulator =A_m*np.sin(2 * np.pi*f_m*t)
product = A_c*np.sin((2*np.pi*f_c*t) + (modulation_index * modulator + 0))

plt.subplot(3,1,1)
plt.title('Phase Modulation')
plt.plot(modulator, 'g')
plt.ylabel('Amplitude')
plt.xlabel('Message Signal')

plt.subplot(3,1,2)
plt.plot(carrier, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(product, color = "purple")
plt.ylabel('Amplitude')
plt.xlabel('PM signal')

plt.subplots_adjust(hspace = 1)
plt.rc('font', size = 30)
fig = plt.gcf()
fig.set_size_inches(16,9)
```
